```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import pprint
import csv
fire = "data/forest_fire_1998-2017.csv"
rain = "data/precipitation_new.csv"
fire_df = pd.read_csv(fire)
rain_df = pd.read_csv(rain)
```

```python
fire_df = fire_df.drop(["Date"],axis=1)
fire_df.head()
```

```python
fire_df = fire_df.rename(columns={'Ano':'Year'})
fire_df.head()
```

```python
rain_df.head()
```

```python
rain_df = rain_df.rename(columns={"state":"State", "precipitation":"Precipitation", "month":"Month", "year":"Year"})
rain_df = rain_df.drop(["Unnamed: 1"], axis=1)
rain_df.head()
```

```python
merged_df = pd.merge(rain_df, fire_df, on=['Year', 'Month','State'])
merged_df
```

```python
merged_df.to_csv("data/merged_df.csv")
```
