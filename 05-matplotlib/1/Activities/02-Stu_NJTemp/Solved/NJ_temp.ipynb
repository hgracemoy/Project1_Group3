```python
# Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set x axis to numerical value for month
x_axis = np.arange(1,13,1)
x_axis
```

```python
plt.hlines(40, 1, 13, alpha=0.25)
```

```python
# Average weather temp
points = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
points_C = [(x-32) * 0.56 for x in points]
points_C
```

```python
points_handle, = plt.plot(x_axis, points, marker ='o', color='blue', label="Fahrenheit")
points_C_handle, = plt.plot(x_axis, points_C, marker='^', color='red', label="Celcius")
```

```python
plt.legend(loc="lower right")
```

```python
plt.show()
```
