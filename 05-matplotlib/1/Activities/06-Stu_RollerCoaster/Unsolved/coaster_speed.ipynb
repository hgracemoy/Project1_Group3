```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
time = np.arange(0,130,10)
drop_speed = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
rail_speed = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
```

```python
plt.plot(time, drop_speed, linewidth=0, marker="o", color="blue")
plt.plot(time, rail_speed, linewidth=0, marker="^", color="red")
```

```python
plt.title("Danger Drop and Railgun Speeds")
plt.xlabel("Seconds Within the Ride (0 to 120)")
plt.ylabel("Speed MPH")
```

```python
plt.grid()
```

```python
plt.show()
```
